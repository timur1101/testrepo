```python
import numpy as np                    # математические операции с векторами
import pandas as pd                   # удобный вывод результатов
import json                           # парсинг JSON-ответов
import time                           # паузы между запросами
from getpass import getpass           # безопасный ввод API-ключа
from IPython.display import display   # красивый вывод таблиц

print("✅ Базовые библиотеки импортированы")
```

```python
from sentence_transformers import SentenceTransformer, util

embed_model = SentenceTransformer('all-MiniLM-L6-v2')

print("✅ Модель эмбеддингов загружена")
print(f"Размер вектора: {embed_model.get_sentence_embedding_dimension()} чисел")
```

```python
import wikipediaapi

# Указываем язык и user-agent
wiki = wikipediaapi.Wikipedia(
    language='en',
    user_agent='TimurRAGProject/1.0 (https://github.com/timur1101)'
)

page = wiki.page("Family_Guy")

print("Title:", page.title)
print("Summary:", page.summary[:500])
print("Text length:", len(page.text))

raw_text = page.text  # полный текст статьи
```

```python
# Разбиваем текст на чанки по предложениям
# Это называется chunking — важный шаг в построении RAG
# В реальных системах используют более сложные стратегии разбивки
sentences = [s.strip() for s in raw_text.split('\n') if s.strip()]

print(f"✅ Корпус создан: {len(sentences)} фрагментов")
print()
print("Первые 3 фрагмента:")
for i, s in enumerate(sentences[:3]):
    print(f"  [{i}] {s[:80]}...")
```

```python
import faiss

# Шаг 1: Создаём эмбеддинги для всех фрагментов корпуса
# show_progress_bar=True — показывает прогресс при большом количестве текстов
print("⏳ Создаём эмбеддинги для всех фрагментов...")
corpus_embeddings = embed_model.encode(sentences, show_progress_bar=True)

print(f"\n✅ Эмбеддинги созданы: форма {corpus_embeddings.shape}")
print(f"   {corpus_embeddings.shape[0]} фрагментов × {corpus_embeddings.shape[1]} чисел")
```

```python
# Шаг 2: Строим FAISS-индекс

# dim — размерность векторов (384 для нашей модели)
dim = corpus_embeddings.shape[1]

# IndexFlatL2 — точный поиск по евклидову расстоянию (L2-норма)
# Подходит для небольших датасетов (до ~100k документов)
index = faiss.IndexFlatL2(dim)

# Добавляем эмбеддинги в индекс
# FAISS требует тип float32 — конвертируем из float64
index.add(np.float32(corpus_embeddings))

print(f"✅ FAISS-индекс построен")
print(f"   Количество векторов в индексе: {index.ntotal}")
print(f"   Размерность векторов: {index.d}")
```

```python
# Шаг 3: Функция семантического поиска

def dense_search(query, top_k=3):
    """
    Ищет наиболее релевантные фрагменты через эмбеддинги (Dense Retrieval).

    Параметры:
        query — текстовый запрос
        top_k — сколько результатов вернуть

    Возвращает: DataFrame с результатами и расстояниями
    """
    # Создаём эмбеддинг запроса — тот же процесс что и для документов
    query_embedding = embed_model.encode([query])

    # index.search() возвращает:
    # distances — расстояния до найденных векторов (меньше = ближе)
    # indices   — индексы найденных фрагментов в нашем списке sentences
    distances, indices = index.search(np.float32(query_embedding), top_k)

    # Формируем читаемый результат
    results = []
    for dist, idx in zip(distances[0], indices[0]):
        results.append({
            "fragment": sentences[idx],
            "distance": round(float(dist), 4),
            "index":    idx
        })

    return results


# Тестируем поиск
query = "When did Family Guy first premiere on television??"
print(f"🔍 Запрос: '{query}'")
print()

results = dense_search(query, top_k=3)
for i, r in enumerate(results):
    print(f"  [{i+1}] Расстояние: {r['distance']}")
    print(f"       Фрагмент: {r['fragment']}")
    print()
```

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model_name = "google/flan-t5-large"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

def ask_llm(question, max_len=128):
    inputs = tokenizer(question, return_tensors="pt")
    outputs = model.generate(**inputs, max_length=max_len)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Пример: LLM без контекста
question = "When did Family Guy first premiere on television?"
print("LLM без контекста:", ask_llm(question))
```

```python
def rag_answer(question, top_k=3):
    results = dense_search(question, top_k=top_k)
    context = " ".join([r["fragment"] for r in results])
    prompt = f"Answer the question using the context below.\n\nContext: {context}\n\nQuestion: {question}"
    return ask_llm(prompt, max_len=256)

print("LLM с RAG:", rag_answer("When did Family Guy first premiere on television?"))
```

```python
questions = [
    "When did Family Guy first premiere on television?",
    "Who created Family Guy?",
    "Which characters are part of the Griffin family?",
    "Has Family Guy ever been cancelled?",
    "What awards has Family Guy received?"
]

for q in questions:
    print(f"\n❓ {q}")
    print("LLM без контекста:", ask_llm(q))
    print("LLM с RAG:", rag_answer(q))
```
